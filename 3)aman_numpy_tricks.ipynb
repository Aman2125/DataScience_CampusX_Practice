#np.sort
Return a sorted copy of an array


```python
#code
import numpy as np
a= np.random.randint(1,100,15)
a
```

```python
b = np.random.randint(1,100,24).reshape(6,4)
b
```

```python
np.sort(a) #default ascending sorting
np.sort(a)[::-1] #for descending order sorting
```

```python
np.sort(b) #for row wise sorting
np.sort(b ,axis=0) #for col wise sorting
```

#np.concatenate
numpy.concatenate() function concatenate a sequence of arrays along an existing axis.

```python
c = np.arange(6).reshape(2,3)
d = np.arange(6,12).reshape(2,3)
print(c)
print(d)
```

```python
np.concatenate((c,d),axis=0) #for row wise concatenation
```

```python
np.concatenate((c,d),axis=1) #for col wise concatenation
```

#Set Functions


*   np.union1d
*   np.intersect1d

*   np.setdiff1d
*   np.setxor1d

*   np.in1d






```python
m = np.array([1,2,3,4,5])
n = np.array([3,4,5,6,7])

np.union1d(m , n)
```

```python
np.intersect1d(m,n)
```

```python
np.setdiff1d(m,n)
```

```python
np.setxor1d(m,n) #common removed
```

```python
np.in1d(m,1)
```
