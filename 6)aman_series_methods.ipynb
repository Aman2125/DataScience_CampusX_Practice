```python
import numpy as np
import pandas as pd
```

```python
subs = pd.read_csv('/content/subs.csv').squeeze()
subs.sample()
```

```python
vk = pd.read_csv('/content/kohli_ipl.csv' , index_col='match_no').squeeze()
vk.sample()
```

```python
movies = pd.read_csv('/content/bollywood.csv' , index_col='movie').squeeze()
movies.sample()
```

```python
#between , tells if value exist in a range , output is boolean
vk[vk.between(51,99)].size
```

```python
#clip , clip in a given range
subs.clip(100,200).head()
```

```python
#drop_duplicates , remove the duplicates values
temp = pd.Series([1,1,2,2,2,3,4,4,5])
temp
```

```python
temp.drop_duplicates()
```

```python
temp = pd.Series([1,2,3,np.nan , 5,6,np.nan,8,np.nan,10])
temp
```

```python
temp.size
```

```python
temp.count()
```

```python
#isnull
temp.isnull().sum() #3 missing values
```

```python
#dropna , drop the na values
temp.dropna()
```

```python
#fillna , replace the missing value with the given value eg 0 here
temp.fillna(0)
```

```python
#isin
vk.sample()
```

```python
vk[(vk == 49) | (vk == 99)] #49 and 99 run by vk
```

```python
vk[vk.isin([49,99])] #simple using isin function
```

```python
#apply
movies.apply(lambda x:x.split()[0].upper()).sample()
```

```python
subs.mean()
```

```python
#que is call a bad day if sub<135 in a day and good when sub>135
subs.apply(lambda x :'good day' if x >135 in subs else 'bad day').sample()
```
