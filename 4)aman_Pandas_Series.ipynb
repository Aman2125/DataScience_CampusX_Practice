#What is Pandas?
Pandas is a fast,powerful,flexible and easy to use open source data analysis and manipulation tool,built on top of the python programming language.



#Pandas Series
A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.



#Importing Pandas

```python
import numpy as np
import pandas as pd
```

#Making Series from Lists

```python
#string
country = ['India' , 'Nepal' ,'Bhutan' , 'Kazakastan' ]
pd.Series(country) #output is index and its value
```

```python
#integer
runs = [13,44,23,66,76,87,89]
pd.Series(runs)
```

```python
#custom index
marks =[56,87,97,54]
subjects = ['English' , 'Maths' , 'CS' , 'Sanskrit']
pd.Series(marks , index = subjects)
```

```python
#setting a name to series
marks =pd.Series(marks , index = subjects , name='Aman ke marks')
marks
```

#Series from Dict

```python
marks ={
    'maths' : 98,
    'Science': 87,
    'English':82,
    'Hindi':88
}

marks_series = pd.Series(marks , name = 'Aman ke marks')
marks_series
```

#Series Attributes

```python
#size , tells size of the series
marks_series.size
```

```python
#dtype , tells the data type of the item in the series
marks_series.dtype
```

```python
#name
marks_series.name
```

```python
#is_unique , check all vale have unique el or not
marks_series.is_unique
```

```python
#index , gives all the index element
marks_series.index
```

```python
#values , gives all the values(numpy array)
marks_series.values
```

#Series using read_csv

```python
#with one col
subs = pd.read_csv('/content/subs.csv').squeeze()  #stores in form of data frame not in series you can check by adding type
#to check in form of series we use squeeze function
subs
```

```python
movies = pd.read_csv('/content/bollywood.csv' , index_col = 'movie').squeeze()
movies
```

#Series Methods

```python
#head and tail , give preview of head and tail of the data
subs.head() #by default give top 5 row
```

```python
vk.head(3) #giving number by which we want
```

```python
vk.tail() #gives last 5 row data by default
```

```python
#sample , randomly gives one row from the data
movies.sample()
#movies.sample(5) , for giving 5 random rows
```

```python
#value_counts -> movies
movies.value_counts() # har actor kitna movie kiya bata dega that is the frequency of the values
```

```python
# sort_index -> inplace -> movies
movies.sort_index()
```

#Series Math Methods

```python
#count , count the total number of items just like size but it doesnt count the missing values
vk.count()
```

```python
#sum -> product
subs.sum() #total sum
subs.product()
```

```python
#mean -> median -> mode -> std -> var
subs.mean()
print(vk.median())
print(movies.mode())
print(subs.std())
print(subs.var())
```

```python
#min /max
subs.min()
subs.max()
```

```python
#describe , give the summary of multiple things
vk.describe()
```
